```python
# Assignment 1
'''
Import videogamesales.csv and perform the following actions on it
Add a column of 'global_sales' showing the total sales of all the different regions to the data frame and sort (highest first) and print the DataFrame according to it.
Display a plot of the total number of copies sold of each genre globally.
Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
a. their name
b. the platform they were released on
c. the year they were released in
d. the sum of sales in only Europe and Japan
Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
'''
```

```python
df=pd.read_csv("videogamesales.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df['global_sales']=df['NA_Sales']+df['EU_Sales']+df['JP_Sales']+df['Other_Sales']
```

```python
df.head()
```

```python
# Display a plot of the total number of copies sold of each genre globally.
copies = df['Genre'].value_counts()
copies
```

```python
plt.figure(figsize=(12,6))
plt.bar(copies.index, copies)
plt.xlabel('Index')
plt.ylabel('Global Sales')
plt.title('Global Sales Bar Chart')
plt.show()
```

```python
# Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
# a. their name
# b. the platform they were released on
# c. the year they were released in
# d. the sum of sales in only Europe and Japan


filtered_df = df[df['Name'].str.contains("Grand Theft Auto", case=False, na=False)]
filtered_df
```

```python
filtered_df = filtered_df[['Name', 'Platform', 'Year', 'EU_Sales', 'JP_Sales']]
```

```python
filtered_df['EU_JP_Sales'] = filtered_df['EU_Sales'] + filtered_df['JP_Sales']
```

```python
filtered_df[['Name', 'Platform', 'Year', 'EU_JP_Sales']]
```

```python
# Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
pie_df =  df[df['Name'].str.contains("Grand Theft Auto", case=False, na=False)]
pie_df
```

```python
total_NA = pie_df['NA_Sales'].sum()
total_JP = pie_df['JP_Sales'].sum()
total_OT = pie_df['Other_Sales'].sum()
total_EU = pie_df['EU_Sales'].sum()
```

```python
sales = [float(total_NA), float(total_EU), float(total_JP), float(total_OT)]
labels = ['North America', 'Europe', 'Japan', 'Other']
explode = [0.05 if s == max(sales) else 0 for s in sales]

plt.figure(figsize=(7,7))
plt.pie(sales,
        labels=labels,
        autopct='%1.1f%%',
        startangle=140,
        explode=explode,
        wedgeprops={'edgecolor':'w'})  # white edges look nicer
plt.title("Total GTA Sales by Region")
plt.axis('equal')   # keeps the pie circular
plt.show()
```
